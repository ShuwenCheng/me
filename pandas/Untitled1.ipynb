```python
categorical = ['AREA NAME','Vict Sex','Status','Status Desc']
continuous = ['TIME OCC','AREA','Rpt Dist No','Part 1-2','Crm Cd','Crm Cd 1','Crm Cd 2','Crm Cd 3','Crm Cd 4']
continuous_time = ['TIME OCC','Rpt Dist No','Crm Cd 1','Crm Cd 2','Crm Cd 3','Crm Cd 4']
```

```python
# turn lower
def normalize_case(s):
    return s.lower()
```

```python
df = df.dropna(subset=['LAT', 'LON'])

latitude = df['LAT']
longitude = df['LON']

map_center = [latitude.mean(), longitude.mean()]  # 设置地图中心为坐标的平均值

m = folium.Map(location=map_center, zoom_start=10)

# 设置实心红点
for lat, lon in zip(latitude, longitude):
    folium.CircleMarker([lat, lon], radius=5, color='red', fill=True, fill_color='red').add_to(m)

m.save('map.html') 
m
```

```python
for column_name in df.columns:
    print('')
    print('')
    #Name
    print(f"Column: {column_name}")
    #What the column describes
    print(f"Describes: this is {column_name} describes")

    if column_name in categorical:
        # How that data was measured
        print("Data Type: Categorical")
        # do a and get an idea of the counts that you'll be working with
        value_counts = df[column_name].value_counts()
        print("Value Counts:", value_counts)
        value_counts.plot(kind="bar")
        plt.title(f"Value Counts of {column_name}")
        plt.savefig(f'{file_path}{column_name}-value-counts.png') 
        plt.show()

        #Check for things that you might need to fold into each other. Do you have entries for and and and in your data? Shouldthey really be the same thing?
        merged_counts = {}
        for item, count in value_counts.items():
            normalized_item = normalize_case(item)
            if normalized_item in merged_counts:
                merged_counts[normalized_item] += count
            else:
                merged_counts[normalized_item] = count

        print(merged_counts)
        merged_counts = pd.Series(merged_counts) 
        merged_counts.plot(kind="bar")
        plt.title(f"Merge Value Counts of {column_name}")
        plt.savefig(f'{file_path}{column_name}-merge-value-counts.png') 
        plt.show()

     
    elif column_name in continuous :
        print("Data Type: Continuous")
        # do a to get an idea of what your numbers are and how they're distributed
        df[column_name].hist()
        plt.title(f"Histogram of {column_name}")
        plt.savefig(f'{file_path}{column_name}-hist.png')  
        plt.show()
        if column_name in continuous_time :
            print("Is Time Series")
            # 假设您的日期列名为 'Date Rptd'
            df['Year'] = df['Date Rptd'].dt.year
            grouped_df = df.groupby('Year')[column_name].mean()  # 这里假设您是要计算每年的平均值，您可以根据需求修改聚合函数
            grouped_df.plot()
            plt.title(f"Trend of {column_name} by Year")
            plt.xlabel("Year")
            plt.ylabel("Value")
            plt.savefig(f'{file_path}{column_name}-time-year.png')  
            plt.show()
        print("Max:", df[column_name].max())
        print("Min:", df[column_name].min())
        print("Mean:", df[column_name].mean())
        print("Median:", df[column_name].median())
        print("Mode:", df[column_name].mode()[0])
   
```
